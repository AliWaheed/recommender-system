# Classification

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC 
from sklearn.neighbors import KNeighborsClassifier  as KNN
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression



df = pd.read_csv("dataset_comb.csv")
```

```python
#print(df.isnull().sum())
```

```python
df.shape
```

```python
df = df.sample(frac = 1)
```

```python
rf = RandomForestClassifier(n_estimators = 100)

pca = PCA() 

target = df['hostle']

kf = KFold(n_splits = 10)

df = df.drop(['hostle'], axis = 1)

acc_total = 0
for train_index, test_index in kf.split(df):
    
    X_train = df.values[train_index]
    Y_train = target[train_index]
    
    X_test = df.values[test_index]
    Y_test = target[test_index]
    
    X_train = pca.fit_transform(X_train)  
    X_test = pca.transform(X_test) 
    
    
    model_random_forest = rf.fit(X_train,Y_train)
    
    predictions = model_random_forest.predict(X_test)
    
    acc = 100*accuracy_score(Y_test,predictions)
    print("Accuracy = {}%".format(acc))
    
    acc_total = acc_total + acc
    
print('Average accuracy = {}'.format(acc_total/10))
```

```python
sv = SVC(gamma=0.1)

acc_total = 0
for train_index, test_index in kf.split(df):
    
    X_train = df.values[train_index]
    Y_train = target[train_index]
    
    X_test = df.values[test_index]
    Y_test = target[test_index]
    
    X_train = pca.fit_transform(X_train)  
    X_test = pca.transform(X_test) 
    
    model_svm = sv.fit(X_train,Y_train)
    
    predictions = model_svm.predict(X_test)
    
    acc = 100*accuracy_score(Y_test,predictions)
    print("Accuracy = {}%".format(acc))
    
    acc_total = acc_total + acc
    
print('Average accuracy = {}'.format(acc_total/10))
```

```python
mb = GaussianNB()
acc_total = 0
for train_index, test_index in kf.split(df):
    
    X_train = df.values[train_index]
    Y_train = target[train_index]
    
    X_test = df.values[test_index]
    Y_test = target[test_index]
    
    X_train = pca.fit_transform(X_train)  
    X_test = pca.transform(X_test) 
    
    model_knn = mb.fit(X_train,Y_train)
    
    predictions = model_knn.predict(X_test)
    
    acc = 100*accuracy_score(Y_test,predictions)
    print("Accuracy = {}%".format(acc))
    
    acc_total = acc_total + acc
    
print('Average accuracy = {}'.format(acc_total/10))
```

# Clustering

```python
import pandas as pd 
from sklearn.cluster import KMeans
import numpy as np

df = pd.read_csv('dataset_comb.csv')

df = df.drop(['hostle'], axis = 1)

headers = list(df)

df = df.T

df_values = df.values

cluster = KMeans(n_clusters = 4)

kmeans = cluster.fit(df_values)


fd = cluster.fit_predict(df_values)

centroids = kmeans.cluster_centers_


df.insert(0, 'Cluster', fd)
```

```python
answer = 'y'

print("--- Enter your two food choices or type 'exit' to leave ---\n")
while (1):
    
    print("\n\nEnter first food item: ")
    answer_one = input()
    
    if answer_one == 'exit':
        break
    
    print("Enter second food item: ")
    answer_two = input()
    
    if answer_two == 'exit':
        break
        
    check = True
        
    if answer_one in cluster_one:
        if answer_two in cluster_one:
            
            check = False
            count = 0
            for i in range(len(cluster_one)):
                if cluster_one[i] in all_frequent_items and cluster_one[i] != answer_one and cluster_one[i] != answer_two:
                    
                    print('I recommend you: {}'.format(cluster_one[i]))
                    break
                count += 1
            
            
            if count == len(cluster_one):
                print('I recommend you: {}'.format(random.choice(cluster_one))) 
        
        if check == False:
                continue
                    
    if answer_one in cluster_two:
        if answer_two in cluster_two:
            
            check = False
            count = 0
            for i in range(len(cluster_two)):
                if cluster_two[i] in all_frequent_items and cluster_two[i] != answer_one and cluster_two[i] != answer_two:
                    
                    print('I recommend you: {}'.format(cluster_two[i]))
                    break
                
                count += 1
            
            if count == len(cluster_two):
                print('I recommend you: {}'.format(random.choice(cluster_two))) 
        
        if check == False:
                continue
    
    if answer_one in cluster_three:
        if answer_two in cluster_three:
            
            check = False
            count = 0
            for i in range(len(cluster_three)):
                if cluster_three[i] in all_frequent_items and cluster_three[i] != answer_one and cluster_three[i] != answer_two:
                    
                    print('I recommend you: {}'.format(cluster_three[i]))
                    break
                count += 1
            if count == len(cluster_two):
                print('I recommend you: {}'.format(random.choice(cluster_three))) 
        
        if check == False:
            continue
    
    else:
        print('I recommend you: {}'.format(random.choice(all_frequent_items)))
```
